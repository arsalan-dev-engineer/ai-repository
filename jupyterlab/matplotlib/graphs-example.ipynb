```python
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
subjects = ['Math', 'Science', 'English', 'History']
grades = {
    'Math': [88, 72, 95, 68, 85],
    'Science': [92, 78, 88, 74, 90],
    'English': [85, 82, 90, 70, 88],
    'History': [80, 75, 85, 82, 78]
}
study_hours = [10, 8, 12, 7, 9]  # Number of hours studied by each student
```

```python
# Bar Chart â€“ Compare Grades by Subject
import matplotlib.pyplot as plt

avg_grades = [sum(grades[subject]) / len(students) for subject in subjects]

plt.bar(subjects, avg_grades, color='skyblue')
plt.title("Average Grades by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Grade")
plt.show()
```

```python
# Line Plot â€“ Track Performance Over Time
# Simulate grades over 5 tests for one student (Alice)
tests = [1, 2, 3, 4, 5]
alice_grades = [75, 80, 85, 78, 88]

plt.plot(tests, alice_grades, marker='o', linestyle='-', color='blue', label='Alice')
plt.title("Alice's Performance Over Time")
plt.xlabel("Test Number")
plt.ylabel("Grade")
plt.legend()
plt.show()
```

```python
# Scatter Plot â€“ Study Time vs. Grades in Math
plt.scatter(study_hours, grades['Math'], color='purple')
plt.title("Study Time vs. Math Grades")
plt.xlabel("Hours Studied")
plt.ylabel("Math Grade")
plt.show()
```

```python
# Histogram â€“ Distribution of Science Grades
plt.hist(grades['Science'], bins=5, color='green', edgecolor='black')
plt.title("Distribution of Science Grades")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Box Plot â€“ Compare Grades Across Students
all_grades = [grades['Math'], grades['Science'], grades['English'], grades['History']]

plt.boxplot(all_grades, labels=subjects)
plt.title("Grades by Subject")
plt.ylabel("Grade")
plt.show()
```

```python
# Pie Chart â€“ Subject Contribution to Total Grade for One Student (Alice)
alice_total = [grades[subject][0] for subject in subjects]

plt.pie(alice_total, labels=subjects, autopct='%1.1f%%', startangle=140)
plt.title("Alice's Grades Contribution by Subject")
plt.show()
```
